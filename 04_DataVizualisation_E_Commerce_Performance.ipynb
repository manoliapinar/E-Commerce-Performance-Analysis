```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
ecommerce_data = pd.read_csv('/content/drive/My Drive/e-commerce-performance-analysis/e-commerce_clickstream_transactions.csv')
clickstream_data = pd.read_csv('/content/drive/My Drive/e-commerce-performance-analysis/e-commerce-dataset.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(clickstream_data['Annual Income'], bins=30, kde=True)
plt.title('Distribution du revenu annuel')
plt.xlabel('Revenu annuel')
plt.ylabel('Fréquence')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=clickstream_data, x='Gender')
plt.title('Répartition des clients par genre')
plt.xlabel('Genre')
plt.ylabel('Nombre de clients')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=clickstream_data, x='Age', y='Annual Income', hue='Gender', alpha=0.6)
plt.title('Relation entre l\'âge et le revenu annuel')
plt.xlabel('Âge')
plt.ylabel('Revenu annuel')
plt.legend(title='Genre')
plt.show()
```

L'analyse des données e-commerce  m'a permis d'explorer des informations clés sur le comportement des clients et leur relation avec les revenus. Les visualisations réalisées ont révélé des tendances intéressantes :

**Distribution du revenu annuel :**
- La distribution des revenus annuels montre des pics à certains niveaux de revenus, ce qui peut indiquer des segments de clients spécifiques qui pourraient être ciblés par des campagnes marketing.

**Répartition par genre :**
- La répartition des clients par genre met en lumière les différences dans la base de clientèle, ce qui peut aider à orienter les stratégies de marketing et de produit pour mieux répondre aux attentes de chaque segment.

**La relation entre l'âge et le revenu annuel :**
- La visualisation de la relation entre l'âge et le revenu annuel suggère qu'il existe des segments d'âge spécifiques avec des revenus potentiellement plus élevés pouvant influencer les décisions de développement de produit et de ciblage publicitaire.

En conclusion, ces insights peuvent aider l'entreprise **à affiner ses stratégies marketing, à personnaliser les offres pour différents segments de clients et à améliorer l'expérience utilisateur.**
